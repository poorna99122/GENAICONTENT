```python
words=["eating","eats","eaten","writing","writes","programming","programs","history","finally","finalized"]
```

```python
## PorterStemmer importing from nltk
from nltk.stem import PorterStemmer
```

```python
## Creating an Object
stemming = PorterStemmer()
```

```python
## We are iterating and getting stem word using stemmming.stem() method 
for word in words:
    print(word + "--->" +stemming.stem(word))
```

```python
## Here in the above stemming process we can observe history - histori which is not the correct strem word 
```

```python
## Here this is the major disadvantage of stemming when stemming is basically appllied for some of the words we may not get correct exact meaning 
## The form of specific word may change
```

```python
stemming.stem('congratulations')
```

```python
## As you can see in the above ... the meaning of the word is actually changing ..
```

```python
## Similarly 
stemming.stem("ssitting")
```

```python
stemming.stem('sitting')
```

```python
## Stemming works for most of the words but for few it wasn't 
```

```python
## RegexpStemmer Class 
from nltk.stem import RegexpStemmer
```

```python
reg_stemmer = RegexpStemmer('ing$|s$|e$|able$', min=4)
```

```python
reg_stemmer.stem('eating')
```

```python
## In short the above reg exp is saying that wherever we have ing or s or e or able at last ... it should get removed
```

```python
reg_stemmer.stem('ingeating')
```

```python
## we should see the reg exp says only last as we have dollor after ing -> ing$
```

```python
#Now if we remove dollor and check The output of above regex will be -> eat
```
