```python
paragraph = """I have three visions for India. In 3000 years of our history, people from all over 
               the world have come and invaded us, captured our lands, conquered our minds. 
               From Alexander onwards, the Greeks, the Turks, the Moguls, the Portuguese, the British,
               the French, the Dutch, all of them came and looted us, took over what was ours. 
               Yet we have not done this to any other nation. We have not conquered anyone. 
               We have not grabbed their land, their culture, 
               their history and tried to enforce our way of life on them. 
               Why? Because we respect the freedom of others.That is why my 
               first vision is that of freedom. I believe that India got its first vision of 
               this in 1857, when we started the War of Independence. It is this freedom that
               we must protect and nurture and build on. If we are not free, no one will respect us.
               My second vision for India’s development. For fifty years we have been a developing nation.
               It is time we see ourselves as a developed nation. We are among the top 5 nations of the world
               in terms of GDP. We have a 10 percent growth rate in most areas. Our poverty levels are falling.
               Our achievements are being globally recognised today. Yet we lack the self-confidence to
               see ourselves as a developed nation, self-reliant and self-assured. Isn’t this incorrect?
               I have a third vision. India must stand up to the world. Because I believe that unless India 
               stands up to the world, no one will respect us. Only strength respects strength. We must be 
               strong not only as a military power but also as an economic power. Both must go hand-in-hand. 
               My good fortune was to have worked with three great minds. Dr. Vikram Sarabhai of the Dept. of 
               space, Professor Satish Dhawan, who succeeded him and Dr. Brahm Prakash, father of nuclear material.
               I was lucky to have worked with all three of them closely and consider this the great opportunity of my life. 
               I see four milestones in my career"""
```

```python
## Will see how to apply stop words 
```

```python
from nltk.stem import PorterStemmer
```

```python
## Here we need to import stop words also
```

```python
from nltk.corpus import stopwords
```

```python
## Now as we imported we also have to download stopwords 
```

```python
import nltk
nltk.download('stopwords')
```

```python
## Currently we have downloaded all language stopwords which is present in NLTK
```

```python
## Now lets see what are the stop words that are available in english
```

```python
stopwords.words('english')
```

```python
stopwords.words('french')
```

```python
stopwords.words('arabic')
```

```python
## Now our sentence is already in english now we will appy stemming and before applying stemming we will remove stopwords from it
```

```python
from nltk.stem import PorterStemmer
```

```python
stemmer = PorterStemmer()
```

```python
sentences = nltk.sent_tokenize(paragraph)
```

```python
type(sentences)
```

```python
## Now we are gonna traverse through all the sentences 
```

```python

## Apply Stopwords And Filter And then Apply Stemming
## Here we are getting start to end iteration and index (Iterating each and every sentence)
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
 ## And from that List of words we are iterating and checking any Stopwords and from that if stopword is not present we are doing stemming
    ## And After stemming we are storing inside words variable 
    words=[stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
 ## And we are converting all the list of words into sentences 
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
sentences
```

```python
## Whatever stopwords that got removed and applied stemming 
```

```python
## We can see this stemming is not proper 
```

```python
## AS we know we can use Snowballl Stemmer for better efficiency 
```

```python
from nltk.stem import SnowballStemmer
snowballstemmer = SnowballStemmer('english')
```

```python
for i in range(len(sentences)):
    words=nltk.word_tokenize(sentences[i])
    words=[snowballstemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
    sentences[i]=' '.join(words)# converting all the list of words into sentences
```

```python
sentences
```

```python
## Here also we can see there is no much change 
```

```python
## we will use lemmatizer for better one again 
```

```python
sent = nltk.sent_tokenize(paragraph)
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
for i in range(len(sent)):
    sentences[i]=sentences[i].lower()
    words=nltk.word_tokenize(sent[i])
    words=[lemmatizer.lemmatize(word.lower(),pos='v') for word in words if word not in set(stopwords.words('english'))]
    sent[i]=' '.join(words)# converting all the list of words into sentences
```
